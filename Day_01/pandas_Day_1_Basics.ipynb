## Pandas
Pandas is a Python library used for **data analysis and data manipulation**.  
It is built on top of **NumPy** and is widely used to work with **structured data**.


```python
import pandas as pd
```

## Series
A **Series** is a **one-dimensional** labeled data structure.  
It is similar to a **single column** in a table.

**Key Points:**
- Contains values and an index
- Supports label-based access
- Can store different data types


```python
# series method 

series = pd.Series([1,2,3,4,5])
series
```

```python
#Series method with index 

series_1 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'])
series_1
```

```python
# Accessing the series element 

print(series[2])
print(series_1['a'])
```

# Pandas DataFrames

### pandas dataframe is like multi dimentional array in numpy it consists of rows as columns

##  DataFrame
A **DataFrame** is a **two-dimensional** data structure with **rows and columns**.  
Each column in a DataFrame is a **Series**.

**Key Points:**
- Similar to an Excel sheet or SQL table
- Can store heterogeneous data
- Most commonly used Pandas object


```python
# DataFrame on a list : 

df_list = pd.DataFrame([1,2,3,4,5])
df_list.head()
```

```python
# DataFrame Using dictionary : 

df = pd.DataFrame({"name" : ["Nikhil","Nitin","Aditya"],"age" : [17,19,21]})
df.head()
```

## Index
An **Index** is used to label rows in Pandas.  
It enables fast data access and alignment.


```python
# DataFrame with index : 

df = pd.DataFrame({"name" : ["Nikhil","Nitin","Aditya"],"age" : [17,19,21]},index=['001','002','003'])
df.head()
```
