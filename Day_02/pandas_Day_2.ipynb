# Pandas DAY 2 

```python
import pandas as pd
```

```python
# Creating a Dataframe : 

student_data = pd.DataFrame({"Name":["Nikhil","Nitin","Aditya"],
                             "Age" : [17,19,21],
                             "University" : ['BHU','DU','LU']},
                             index=['001','002','003'])
```

```python
student_data
```

## Inspecting Data (Shape, Columns , Data_type)

```python
# To check the shape : 
print("Shape of dataset is : ")
print(student_data.shape)


# To check the Columns : 
print("\nColumns of dataset are :")
print(student_data.columns)


#To check the datatypes : 
print("\nDatatype of columns are : ")
print(student_data.dtypes)
```

## Identifying Numerical and Categorical Columns

Numerical columns represent **quantities** on which mathematical operations (mean, sum, etc.) make sense, such as age or salary.

Categorical columns represent **labels or groups** used for classification, such as gender, department, or city.

### How to identify
- Columns with `int` or `float` data types are usually **numerical**
- Columns with `object` or `category` data types are usually **categorical**
- Some numeric-looking columns (like `id`) are identifiers and should be treated as **categorical**

### Key Point
Data type helps identify column type, but the **meaning of the data** is more important.


```python

# Identify numerical columns
numerical_cols = student_data.select_dtypes(include=["int64", "float64"]).columns

# Identify categorical columns
categorical_cols = student_data.select_dtypes(include=["object"]).columns

print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

## Column selection 

Column selection is used to choose specific columns from a DataFrame.

- Selecting a single column returns a Series  
- Selecting multiple columns returns a DataFrame


```python
# Selecting the Name column from student_data dataset
student_data['Name']
```

## Row Filtering

Row filtering is used to select rows based on a condition.

- Conditions are applied on column values  
- Only rows that satisfy the condition are returned


```python
# Filtering rows using condition
student_data[student_data["Age"] > 18]
```


## loc vs iloc
- **loc** → Label-based indexing  
- **iloc** → Integer position-based indexing  

Use **loc** with row/column names.  
Use **iloc** with numerical positions.

```python
# Row selection 

student_data.loc["001",'Name']
student_data.loc["003",'Age']


# Row selection by index using iloc

student_data.iloc[1,1]
student_data.iloc[2,0]
```
